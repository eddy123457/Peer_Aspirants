```python

#this is tarleton information for each KPI I found all my bugs using this method of search; if some dtypes were messed up they would show up as NANS so thanks to tarleton for always reporting 
import pandas as pd


master_table['unitid'] = master_table['unitid'].astype(str)
master_table['Metric'] = pd.to_numeric(master_table['Metric'], errors='coerce')

kpi_tables = {}

unique_kpis = master_table['KPI_Description'].unique()
unitid_filter = '228529'
#filter each KPI then group and sort
for kpi in unique_kpis:
    filtered_data_for_kpi = master_table[
        (master_table['unitid'] == (unitid_filter)) &
        (master_table['KPI_Description'] == kpi)
    ][['unitid', 'Year', 'KPI_Description','Goal', 'Metric']].sort_values(by='Year')
    
    kpi_tables[kpi] = filtered_data_for_kpi
    #show what filtered table 
    print(f"\nFiltered Data for KPI: {kpi}")
    display(filtered_data_for_kpi)  


```

```python
#all these PD.conversion (numeric or string types) are because in some tables in the goal notebooks there were mixed types that when read obv show object types from pandas
#didnt think anything of it but for instance in metric column some schools would report with COMMMAS and other with no commas in fact only like 30ish schools used commas and everyone else didnt sooo not something
#i wouldve caught but tarleton was one of those schools so for every KPI i checked for commas and my NANS went down by alot 
#we can remove them as i think they arent needed but the code works with them in so i just left thm here

master_table['Metric'] = pd.to_numeric(master_table['Metric'], errors='coerce')

unique_unitids = master_table['unitid'].unique()
unique_kpis = master_table['KPI_Description'].unique()

interpolated_results = pd.DataFrame()

total_nans_before = master_table['Metric'].isna().sum()
#interpolating by KPI_Description and not whole set
for unitid in unique_unitids:
    for kpi in unique_kpis:
        filtered_data_for_kpi = master_table[
            (master_table['unitid'] == unitid) &
            (master_table['KPI_Description'] == kpi)
        #order by year for inter 
        ][['unitid', 'Year', 'KPI_Description', 'Metric', 'Goal']].sort_values(by='Year')
        #this tracks where NANS existed
        filtered_data_for_kpi['NaNs_Before'] = filtered_data_for_kpi['Metric'].isna().sum()
        
        #ended with this since the totals NANS went from 28000 to 292 so we can actually use it and it have little no impact but since we are filtering for each KPI_Description if we impute here it will only impute from values associated with 
        #this school and this KPI; as before we thought it would impute with values learned from other schools; but what we have here will work 
        filtered_data_for_kpi['Metric'] = filtered_data_for_kpi['Metric'].interpolate(method='linear').ffill().bfill()
        
        filtered_data_for_kpi['NaNs_After'] = filtered_data_for_kpi['Metric'].isna().sum()
        
        interpolated_results = pd.concat([interpolated_results, filtered_data_for_kpi])

total_nans_after = interpolated_results['Metric'].isna().sum()

nan_rows = interpolated_results[interpolated_results['Metric'].isna()]

nan_details = nan_rows[['unitid', 'Year', 'KPI_Description', 'Metric', 'Goal']]

print(f"Total NaNs Before: {total_nans_before}, Total NaNs After: {total_nans_after}")
display(nan_details)

#These UNITIDS 214713,498571 will be dropped as they did not report (Pennsylvania State University-Penn State Harri,Pennsylvania Western University	)-> only for 2022 aactually just fill zero 
#interpolated_results = interpolated_results.dropna(subset=['Metric'])

#meant to drop these but since they report for other goals then it made a mess so i just filled these with zero so we dont get nans but here are the display nans 
interpolated_results['Metric'] = interpolated_results['Metric'].fillna(0)
```

```python

#this is where i merge these files notice i use the suffix route so we can track the interpolated values though i do drop later but we dont have to 
master_table.reset_index(inplace=True)
interpolated_results.reset_index(inplace=True)

updated_master_table = master_table.merge(interpolated_results, on=['unitid', 'Year', 'KPI_Description'], how='left', suffixes=('', '_interpolated'))


display(updated_master_table)
#update the table with the values interpolated from the metric column I couldnt get the .update to work here but it simulates that
updated_master_table['Metric'] = updated_master_table['Metric'].combine_first(updated_master_table['Metric_interpolated'])

updated_master_table.drop(columns=['Metric_interpolated','index_interpolated'], inplace=True)

total_nans_after_update = updated_master_table['Metric'].isna().sum()
display(total_nans_after_update)
display(updated_master_table)
display(total_nans_after_update)

```

```python
#final PCA calulation This one is a using the KPI Target get devation normilize the weights to that devation  then subtract the reference 'tarleton' to ranlk
from sklearn.decomposition import PCA
from sklearn.preprocessing import RobustScaler
import pandas as pd

weights = {
    '1': 3,
    '2': 4,
    '3': 2,
    '4': 4,
    '5': 1,
}
#norm the weights
total_weight = sum(weights.values())
weights = {k: v / total_weight for k, v in weights.items()}
#note that i mean by KPI and not goal or objective bc some of these are mixed with percentages and it wouldnt make sense to do it as an avg by obj or goal
grouped = updated_master_table.groupby(['unitid', 'KPI_Description']).agg({
    'Metric': 'mean',
    'KPI_Target': 'mean',
    'Goal': 'first', #this just makes sure that its included in the final table
    'INSTNM': 'first'  
}).reset_index()

#how far is the the school from the target
grouped['Deviation'] = grouped['Metric'] - grouped['KPI_Target']
#display(grouped.head())
#apply weights to that
grouped['Normalized_Weight'] = grouped['Goal'].astype(str).map(weights)

grouped['Weighted_Deviation'] = grouped['Deviation'] * grouped['Normalized_Weight']
#display(grouped.head())

scaler = RobustScaler()

pca_results = []
#preform pca per each KPI_Description
for kpi in grouped['KPI_Description'].unique():
    kpi_data = grouped[grouped['KPI_Description'] == kpi][['unitid', 'Weighted_Deviation']].set_index('unitid')
    
    standardized_data = scaler.fit_transform(kpi_data)
    
    pca = PCA(n_components=1)
    pca_result = pca.fit_transform(standardized_data)
    
    pca_df = grouped[grouped['KPI_Description'] == kpi].copy()
    pca_df['PCA_Value'] = pca_result.flatten()
    pca_results.append(pca_df)

pca_combined = pd.concat(pca_results, ignore_index=True)

reference_unitid = '228529'

#make sure that it exist
if reference_unitid not in pca_combined['unitid'].unique():
    raise ValueError(f"Reference unitid {reference_unitid} not found in the PCA results.")

comparison_results = []
#gouping values and grabbing P and A per KPI_
for (kpi_desc, goal), group_df in pca_combined.groupby(['KPI_Description', 'Goal']):
    ref_df = group_df[group_df['unitid'] == reference_unitid]
    if ref_df.empty:
        continue
    
    ref_pca_value = ref_df['PCA_Value'].iloc[0]
    
    group_df = group_df.copy()
    group_df['Difference'] = group_df['PCA_Value'] - ref_pca_value
    
    comp_df = group_df[group_df['unitid'] != reference_unitid].copy()
    
    positive_diff = comp_df[comp_df['Difference'] > 0].sort_values('Difference', ascending=True)
    negative_diff = comp_df[comp_df['Difference'] < 0].sort_values('Difference', ascending=False)

    top_5_aspirants = positive_diff.head(5).copy()
    top_5_aspirants['Type'] = 'Aspirant'
    
    top_5_peers = negative_diff.head(5).copy()
    top_5_peers['Type'] = 'Peer'
    
    combined_results = pd.concat([top_5_aspirants, top_5_peers], ignore_index=True)
    combined_results['Reference_unitid'] = reference_unitid
    combined_results['Reference_PCA_Value'] = ref_pca_value
    combined_results['KPI_Description'] = kpi_desc
    combined_results['Goal'] = goal
    
    comparison_results.append(combined_results)

final_comparison_df = pd.concat(comparison_results, ignore_index=True)
#since i had no values in the tables for this number i just left as zero in the KPI_taget so these schools are stil P and A the KPI_target is at zero 
#final_comparison_df = final_comparison_df[final_comparison_df['KPI_Description'] != 'sponsored research']

final_comparison_df = final_comparison_df[['unitid', 'INSTNM', 'Goal', 'PCA_Value', 'Type', 'Difference', 'KPI_Description','Goal']]

final_comparison_df
```

```python
#same code but this is another version of this effectly answering how do schools measure to the reference unitid and not a target_kpi but comparivly speaking since i assume there isnt any change to the variance pattern after standardizing the data it gives the same value for tarleton


from sklearn.decomposition import PCA
from sklearn.preprocessing import RobustScaler
import pandas as pd

weights = {
    '1': 3,
    '2': 4,
    '3': 2,
    '4': 4,
    '5': 1,
}

total_weight = sum(weights.values())
weights = {k: v / total_weight for k, v in weights.items()}

reference_unitid = '228529'

grouped = updated_master_table.groupby(['unitid', 'KPI_Description']).agg({
    'Metric': 'mean',
    'Goal': 'first',
    'INSTNM': 'first',
    'Objective': 'first'  
}).reset_index()

grouped['Normalized_Weight'] = grouped['Goal'].astype(str).map(weights)
display(grouped.head())
pca_results = []

scaler = RobustScaler()

comparison_results = []

for (kpi_desc, goal), group_df in grouped.groupby(['KPI_Description', 'Goal']):
    ref_row = group_df[group_df['unitid'] == reference_unitid]
    if ref_row.empty:
        continue
    
    reference_metric = ref_row['Metric'].iloc[0]

    group_df['Difference'] = group_df['Metric'] - reference_metric

    group_df['Weighted_Difference'] = group_df['Difference'] * group_df['Normalized_Weight']

    kpi_data = group_df[['unitid', 'Weighted_Difference']].set_index('unitid')
    standardized_data = scaler.fit_transform(kpi_data)

    pca = PCA(n_components=1)
    pca_result = pca.fit_transform(standardized_data)

    group_df['PCA_Value'] = pca_result.flatten()
    group_df['Reference_unitid'] = reference_unitid
    group_df['Reference_Metric'] = reference_metric
    group_df['KPI_Description'] = kpi_desc
    group_df['Goal'] = goal

    comp_df = group_df[group_df['unitid'] != reference_unitid].copy()

    ref_pca_value = group_df.loc[group_df['unitid'] == reference_unitid, 'PCA_Value'].iloc[0]
    comp_df['Difference_from_Ref_PCA'] = comp_df['PCA_Value'] - ref_pca_value

    positive_diff = comp_df[comp_df['Difference_from_Ref_PCA'] > 0].sort_values('Difference_from_Ref_PCA', ascending=True)
    negative_diff = comp_df[comp_df['Difference_from_Ref_PCA'] < 0].sort_values('Difference_from_Ref_PCA', ascending=False)

    top_5_aspirants = positive_diff.head(5).copy()
    top_5_aspirants['Type'] = 'Aspirant'

    top_5_peers = negative_diff.head(5).copy()
    top_5_peers['Type'] = 'Peer'

    final_group = pd.concat([top_5_aspirants, top_5_peers], ignore_index=True)
    final_group['Reference_PCA_Value'] = ref_pca_value
    final_group['Adjusted_PCA_Value'] = final_group['PCA_Value'] - ref_pca_value

    comparison_results.append(final_group)

final_comparison_df = pd.concat(comparison_results, ignore_index=True)

#final_comparison_df = final_comparison_df[final_comparison_df['KPI_Description'] != 'sponsored research']

final_comparison_df = final_comparison_df[
    ['unitid', 'INSTNM', 'Goal', 'Objective', 'Metric', 'PCA_Value', 
     'Reference_unitid', 'Reference_Metric', 'Reference_PCA_Value', 'Adjusted_PCA_Value', 
      'Type', 'KPI_Description']
]

display(final_comparison_df)
```
